```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from numpy import arange
import matplotlib
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import KFold
from math import sqrt
pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
%matplotlib inline

# Converte variáveis numéricas para a base logarítmica
# Pode ser passado por parâmetro uma única variável, ou mesmo o dataframe inteiro.
def converte_para_ln(df,cols=''):
 if cols == '':
   cols = df.columns.tolist()
 contador = 0 
 for col in cols: 
   if df[col].dtype in [np.int64, np.int32, np.float64]:
     nova_coluna = col + "_ln" 
     # A linha abaixo calcula o log de cada valor das colunas numéricas
     # O abs para x é porque o np.log retorna erro para valores negativos, usaremos,
     # A função log retonra um erro (-inf) quando o x = 0
     # Assim, quando o valor for zero, a função retornará -13,82 (que é o ln de 0,00
     df[nova_coluna]=df[col].apply(lambda x: np.log(abs(x)) if x!= 0 else -13,82).astype(float)
     contador = contador + 1
 print ('Colunas criadas:',contador) 
 return('Concluído')

# Converte variáveis numéricas usando a técnica MinMax
def converte_para_MinMax(df,cols): 
 contador = 0 
 for col in cols: 
   #define um novo nome para as colunas transformadas
   nova_coluna = col + "_minmax"
   # transforma os dados para o formato exigido pelo MinMaxScaler() e procede à transf
   dados_a_transformar = df[col].values.reshape(-1, 1)
   scaler = MinMaxScaler()
   dados_transformados = scaler.fit_transform(dados_a_transformar)
   # restabelece o formato DataFrame e atribui o valor à nova coluna criada
   df[nova_coluna] = DataFrame(dados_transformados)
   #print (col, nova_coluna, " - Min", scaler.data_min_, " - Máx", scaler.data_max_, "
   contador = contador + 1
 print ('Colunas criadas:',contador) 
 return('Concluído') 

# Calcula a métrica Root Mean Squared Error
def rmse(y_true, y_pred):
 return sqrt(mean_squared_error(y_true,y_pred))

#Crira um scatter plot com o resultado do modelo de machine learning.
def imprime_scatter_plot(y_test,y_pred):
 plt.scatter(y_test, y_pred)
 range = [y_test.min(), y_pred.max()]
 plt.plot(range, range, 'red')
 plt.xlabel('Demanda real (em log neperiano)')
 plt.ylabel('Demanda predita (em log neperiano)')
 plt.show()
 return ''



#aplica modelos de machine learning
def aplicar_modelo_ML(df, modelo_definido, x_train, y_train, x_test, y_test):
 
 print('Dataframe:',df)
 
 modelo = modelo_definido 
 print("Modelo:",modelo)
 
 modelo.fit(x_train,y_train)
 y_pred = modelo.predict(x_train)
 rmse_train=rmse(y_train,y_pred)
 mae_train = mean_absolute_error(y_train,y_pred)
 print("MAE na base de treinamento: ",mae_train)
 print("RMSE na base de treinamento: ",rmse_train)
 print("R² na base de treinamento: {:.2f}".format(modelo.score(x_train, y_train)))
 
 modelo.fit(x_train,y_train)
 y_pred = modelo.predict(x_test)
 rmse_test=rmse(y_test,y_pred)
 mae_test = mean_absolute_error(y_test,y_pred) 
 print("MAE na base de teste: ",mae_test)
 print("RMSE na base de teste: ",rmse_test)
 print("R² na base de teste: {:.2f}".format(modelo.score(x_test, y_test)))
 print("\n")
 
 y_test_ln = np.log(abs(np.where(y_test==0,0.00001,y_test)))
 y_pred_ln = np.log(abs(np.where(y_pred==0,0.00001,y_pred)))
 
 #imprime_scatter_plot(np.log(y_test),np.log(y_pred))
 imprime_scatter_plot(y_test_ln,y_pred_ln)
 
 resultados = {'Dataframe': df,
               'Modelo': [str(modelo)],
               'MAE - Base Treinamento':mae_train,
               'MAE - Base Teste':mae_test,
               'RMSE - Base Treinamento': rmse_train,
               'RMSE - Base Teste': rmse_test,
               'R2 (Acurácia) - Base Treinamento': modelo.score(x_train, y_train), 
               'R2 (Acurácia) - Base Teste': modelo.score(x_test, y_test)} 
 
 return resultados


def apura_coeficientes(modelo_usado,X,x_train,y_train):
 
 
 modelo = modelo_usado.fit(x_train,y_train)
 nome_coluna = 'Coeficientes_' + str(modelo)[:-2]
 nome_coluna_modulo = 'Coeficientes_' + str(modelo)[:-2] +'_módulo'
 
 coeff = pd.Series(X.columns,name='Variáveis independentes').to_frame()
 coeff[nome_coluna] = pd.Series(modelo.coef_,name='Coeficientes').astype(float).to_frame
 coeff[nome_coluna_modulo] = (coeff[nome_coluna]**2)**0.5
 coeff=coeff.sort_values(by=[nome_coluna_modulo],ascending=False)
 features = coeff['Variáveis independentes']
 coeficientes = coeff[nome_coluna_modulo]
 
 plt.figure(figsize=(10,5))
 plt.ylabel('Coeficientes (em módulo)')
 plt.xlabel('Variáveis independentes')
 plt.title('Variáveis independentes e Coeficientes (em módulo)')
 plt.bar(features,coeficientes)
 plt.xticks(rotation = 90)
 plt.show()
 return coeff


def plotstats(df, col):
 import matplotlib.pyplot as plt
 
 plt.rc('axes', labelsize=20) 
 plt.rc('xtick', labelsize=20) 
 plt.rc('ytick', labelsize=20) 
 
 ## Setup for ploting two charts one over the other
 fig, ax = plt.subplots(2, 1, figsize = (12,8))
 ## First a box plot
 #df.dropna().boxplot(col, ax = ax[0], vert=False,return_type='dict')
 df.boxplot(col, ax = ax[0], vert=False,return_type='dict')
 ## Plot the histogram
 ## temp = df[col].as_matrix()
 temp = df[col].values
 ax[1].hist(temp, bins = 30, alpha = 0.7)
 plt.ylabel('Número de Cidades')
 plt.xlabel(col)
  
 return [col]



def plot_residuo(df,x_indicado,y_indicado,coluna_residuo,predicao,preditor):
 
 nrow = df.shape[0]
 df.sort_values(by=y_indicado, ascending=True, inplace = True)
 fig, ax = plt.subplots(2, 1, figsize = (36,36))
 
 fonte = 40
 
 plt.rc('axes', labelsize=fonte) 
 plt.rc('xtick', labelsize=fonte) 
 plt.rc('ytick', labelsize=fonte)
 
 df.plot(kind = 'scatter', x = x_indicado, y = y_indicado , ax = ax[0], alpha = 0.5)
 
 ax[0].set_xlabel('Municipios (ordem ascendente de Demanda)',size=fonte)
 ax[0].set_ylabel('Resíduos (base logarítmica)', size=fonte)
 #ax[0].set_title('X vs. Y',size=fonte)
 
 #ax[0].scatter(df_log['indice_ordenado'],df_log['01_arrecadacao_2018_ln'],color='black'
 #ax[0].set_xlabel('Municípios (em ordem crescente de Demanda)')
 #ax[0].set_ylabel('Demanda (base logarítmica)')
 
 ax[1].hist(df[coluna_residuo], bins = 30, alpha = 0.7) 
 #ax[1].hist(df['01_arrecadacao_2018_ln'], bins = 30, alpha = 0.7, histtype='step', colo
 ax[1].set_xlabel('Resíduos (base logaritmica)',size=fonte)
 ax[1].set_ylabel('Número de municípios',size=fonte)
 ax[1].set_title('Histograma dos Resíduos',size=fonte) 
 
 return ' '


def plot_arrecadacao(df, faixa_arrec): 
 
 matplotlib.use('agg')
 
 df.sort_values(by='indice_ordenado', ascending=True, inplace = True)
 for tm in faixa_arrec:
   fig = plt.figure(figsize=(50, 10))
   fig.clf()
   ax = fig.gca()
   ### ???
   df[df['faixa_arrecadacao'] == tm].plot(kind = 'line', x = 'indice_ordenado', y = '01_arrecadacao_2018_ln', color='black')
   df[df['faixa_arrecadacao'] == tm].plot(kind = 'line', x = 'indice_ordenado', y = 'predicao_melhor_modelo_ln', color='black')
   fonte = 40
   plt.xlabel("Faixas Demanda", size = fonte)
   plt.ylabel("Demanda", size = fonte)
   plt.title("Demanda por Faixa Demanda = " + str(tm), size = fonte)
   plt.legend(fontsize = fonte/1.2)  
   #fig.savefig('faixa_arrec_' + str(tm) + '.png') 
 return 'Concluído'



def box_residuos(df):
 
 matplotlib.use('agg') # Define backend
 
 #df['residuos_%'] = df['residuos_ln']/df['01_arrecadacao_2018_ln']*100
 df['residuos_%'] = df['residuos']/df['01_arrecadacao_2018']*100
 
 fig = plt.figure(figsize=(12, 6))
 fig.clf()
 ax = fig.gca() 
 ### ???
 df[df['residuos_%'] <= 200].boxplot(column = ['residuos_%'], by = ['faixa_arrecadacao'], ax = ax) 
 #df.boxplot(column = ['residuos_ln'], by = ['faixa_arrecadacao'], ax = ax) 
 
 fonte = 15
 
 plt.xlabel('',size = fonte)
 plt.ylabel('Resíduos_%',size=fonte)
 plt.title('',size=fonte)
 #plt.set_xticklabels(x_ticks, rotation=0, fontsize=fonte)
 #plt.set_yticklabels(y_ticks, rotation=0, fontsize=fonte)
 
 #plt.rc('axes', labelsize=fonte) 
 #plt.rc('xtick', labelsize=fonte) 
 #plt.rc('ytick', labelsize=fonte)
 
 #fig.savefig('boxes' + '.png')
 return 'Concluído'


def resids_hist(df, faixa_arrec):
 
 matplotlib.use('agg') 
 
 plt.rc('axes', labelsize=20) 
 plt.rc('xtick', labelsize=20) 
 plt.rc('ytick', labelsize=20) 
 
 for tm in faixa_arrec:
   fig = plt.figure(figsize=(8, 6))
   fig.clf()
   ax = fig.gca()
   ax.hist(df.loc[df['faixa_arrecadacao'] == tm, 'residuos_ln'].values, bins = 30)
   plt.xlabel("Residuals")
   plt.ylabel("Density")
   plt.title("Histograma dos resíduos por faixa de valor = " + str(tm))
   #fig.savefig('hist_' + str(tm) + '.png')
 
 return 'Concluído'

#1)LER DADOS E PREPARAR OS DATAFRAMES
df_integral_outliers=pd.read_excel('o_dataframe.xlsx', usecols=lambda x: x if not x.startswith('Unnamed') else None)
df_integral=pd.read_excel('o_dataframe_sem_outliers_correlacao_ok.xlsx', usecols=lambda x: x if not x.startswith('Unnamed') else None)
#Excluindo colunas que já foram descartadas em virtude da correlação:
df_integral_outliers = df_integral_outliers.drop(columns=list(set(df_integral_outliers.columns) - set(df_integral.columns)))

print(df_integral.shape)
print(df_integral.dtypes)
df_integral.head()
###MODTESTE
#df_integral['Recebidos_2015_a_2023']=df_integral['Recebidos_2015_a_2023']/df_integral['População residente - pessoas [2022]_005']
#Experiência Outliers 1L
#df_integral = df_integral_outliers
df_original = df_integral.copy()


df_log = df_original.copy()
#Seleciona as colunas numéricas...
colunas_a_converter = df_log.dtypes[df_log.dtypes != "object"].index.tolist()
#... exceto a última: 'Recebidos_2015_a_2023'.
colunas_a_converter.pop()
print(colunas_a_converter)
converte_para_ln(df_log, colunas_a_converter)
df_log = df_log.drop(colunas_a_converter,axis=1)
print (df_log.shape)
df_log.describe()


df_MinMax = df_original.copy()
#Seleciona as colunas numéricas...
colunas_a_converter = df_MinMax.dtypes[df_MinMax.dtypes != "object"].index.tolist()
#... exceto a última
colunas_a_converter.pop()
# Chama a função que fará a conversão
converte_para_MinMax(df_MinMax, colunas_a_converter)
# Apaga as colunas originais
df_MinMax = df_MinMax.drop(colunas_a_converter,axis=1)
print (df_MinMax.shape)
df_MinMax.describe()


#Separação BASE DE TREINAMENTO vs BASE DE TESTE
#Identificar a coluna que contém o Label (Y)
Y_orig=df_original['Recebidos_2015_a_2023']
#X será as demais colunas. Dado X tenho Y.
X_orig=df_original.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
x_orig_train, x_orig_test, y_orig_train, y_orig_test = train_test_split(X_orig,Y_orig,test_size=0.3, random_state=17)

#Identificar a coluna que contém o Label (Y)
Y_minmax=df_MinMax['Recebidos_2015_a_2023']
#X será as demais colunas. Dado X tenho Y.
X_minmax=df_MinMax.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
x_minmax_train, x_minmax_test, y_minmax_train, y_minmax_test = train_test_split(X_minmax,Y_minmax,test_size=0.3, random_state=17)

#Identificar a coluna que contém o Label (Y)
Y_log=df_log['Recebidos_2015_a_2023']
#X será as demais colunas. Dado X tenho Y.
X_log=df_log.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
x_log_train, x_log_test, y_log_train, y_log_test = train_test_split(X_log,Y_log,test_size=0.3, random_state=17)

###INSECURE 001 os 3 acima truncaram em test_size
#273.13 é a média
baseline_orig = mean_absolute_error(y_orig_test, np.full(y_orig_test.shape,df_integral['Recebidos_2015_a_2023'].mean(),dtype=float))
print ('Baseline para o modelo utilizando os dados originais', baseline_orig)
#rmse(y_test, np.zeros(y_test.shape))
baseline_minmax = mean_absolute_error(y_minmax_test, np.full(y_minmax_test.shape,df_integral['Recebidos_2015_a_2023'].mean(),dtype=float))
print ('Baseline para o modelo utilizando os dados transformados - MinMax', baseline_minmax)
#rmse(y_test, np.zeros(y_test.shape))
baseline_log = mean_absolute_error(y_log_test, np.full(y_log_test.shape,df_integral['Recebidos_2015_a_2023'].mean(),dtype=float))
print ('Baseline para o modelo utilizando os dados transformados - Log Natural', baseline_log)
###INSECURE 002 os 3 acima truncaram em np.full
##I - APLICANDO REGRESSAO LINEAR
modelo = LinearRegression()
#Aplicação LinearRegression() no df_original
resultado = aplicar_modelo_ML('df_original', modelo, x_orig_train, y_orig_train, x_orig_test, y_orig_test)
df_resultados_modelos = pd.DataFrame(resultado)
#Aplicação LinearRegression() no df_MinMax
resultado = aplicar_modelo_ML('df_MinMax', modelo, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação LinearRegression() no df_log
resultado = aplicar_modelo_ML('df_log', modelo, x_log_train, y_log_train, x_log_test, y_log_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
###INSECURE 003 os 3 acima truncaram em np.full

##II - APLICANDO RIDGE
modelo = Ridge()
#Aplicação Ridge() no df_original
resultado = aplicar_modelo_ML('df_original', modelo, x_orig_train, y_orig_train, x_orig_test, y_orig_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação Ridge() no df_MinMax
resultado = aplicar_modelo_ML('df_MinMax', modelo, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação Ridge() no df_log
resultado = aplicar_modelo_ML('df_log', modelo, x_log_train, y_log_train, x_log_test, y_log_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)

##III - APLICANDO LASSO
# O parâmetro max_iter foi definido abaixo por conta de um aviso exibido pelo python no mom
modelo = Lasso()
#Aplicação Lasso() no df_original
resultado = aplicar_modelo_ML('df_original', modelo, x_orig_train, y_orig_train, x_orig_test, y_orig_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
#Aplicação Lasso() no df_MinMax
resultado = aplicar_modelo_ML('df_MinMax', modelo, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
#Aplicação Lasso() no df_log
resultado = aplicar_modelo_ML('df_log', modelo, x_log_train, y_log_train, x_log_test, y_log_test)
#df_resultados_modelos = df_resultados_modelos.append(resultado,ignore_index=True)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)

#IV - ELASTIC NET
modelo = ElasticNet()
#Aplicação ElasticNet() no df_original
resultado = aplicar_modelo_ML('df_original', modelo, x_orig_train, y_orig_train, x_orig_test, y_orig_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação ElasticNet() no df_MinMax
resultado = aplicar_modelo_ML('df_MinMax', modelo, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação ElasticNet() no df_log
resultado = aplicar_modelo_ML('df_log', modelo, x_log_train, y_log_train, x_log_test, y_log_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)

#V - RANDOMFOREST
modelo = RandomForestRegressor()
#Aplicação RandomForestRegressor() no df_original
resultado = aplicar_modelo_ML('df_original', modelo, x_orig_train, y_orig_train, x_orig_test, y_orig_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação RandomForestRegressor() no df_MinMax
resultado = aplicar_modelo_ML('df_MinMax', modelo, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)
#Aplicação RandomForestRegressor() no df_log
resultado = aplicar_modelo_ML('df_log', modelo, x_log_train, y_log_train, x_log_test, y_log_test)
df_dictionary = pd.DataFrame([resultado])
df_resultados_modelos = pd.concat([df_resultados_modelos, df_dictionary], ignore_index=True)

#Aponta o melhor modelo.
print ('O melhor resultado foi com o modelo',df_resultados_modelos['Modelo'][df_resultados_modelos['MAE - Base Teste']==min(df_resultados_modelos['MAE - Base Teste'])])
print ('Para o dataframe',df_resultados_modelos['Dataframe'][df_resultados_modelos['MAE - Base Teste']==min(df_resultados_modelos['MAE - Base Teste'])])
print('MAE:',min(df_resultados_modelos['MAE - Base Teste']))
print('RMSE:',min(df_resultados_modelos['RMSE - Base Teste']))
print('R²:',max(df_resultados_modelos['R2 (Acurácia) - Base Teste']))

df_resultados_modelos

###ANALISE DE RESÍDUOS DO MELHOR MODELO
#Isolar as features e o label
Y=df_MinMax['Recebidos_2015_a_2023']
X=df_MinMax.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
#reaplicar o modelo com os melhore hiperparâmetros
modelo = Ridge(alpha=11.12).fit(X,Y)
#modelo = Ridge(alpha=0.1).fit(X,Y)
y_pred = modelo.predict(X)
#cria coluna com as predições
df_MinMax['predicao_melhor_modelo'] = y_pred
mae_total = mean_absolute_error(Y,y_pred)
print("MAE-total", mae_total)
# Calcular os resíduos
df_MinMax['residuos'] =abs(df_MinMax['Recebidos_2015_a_2023'] - df_MinMax['predicao_melhor_modelo'])
df_MinMax['residuos_sinal'] =df_MinMax['Recebidos_2015_a_2023'] - df_MinMax['predicao_melhor_modelo']
print(df_MinMax[['municipio_002','Recebidos_2015_a_2023','predicao_melhor_modelo','residuos_sinal']].sort_values('residuos_sinal').to_string())
print(df_MinMax.columns)

"""
###ANALISE DE RESÍDUOS DO MELHOR MODELO REMOVENDO TEMPO
df_MinMax_Sem_Tempo = df_integral.copy()
colunas_a_converter = df_MinMax_Sem_Tempo.dtypes[df_MinMax_Sem_Tempo.dtypes != "object"].index.tolist()
colunas_a_converter.pop()
converte_para_MinMax(df_MinMax_Sem_Tempo, colunas_a_converter)
df_MinMax_Sem_Tempo = df_MinMax_Sem_Tempo.drop(colunas_a_converter,axis=1)
df_MinMax_Sem_Tempo['tempo_s_minmax'] = 0.1
#Isolar as features e o label
###Isso que tava bugando: df_MinMax_Sem_Tempo = df_MinMax.copy() -- Fazendo ficar tudo milesimal
Y=df_MinMax_Sem_Tempo['Recebidos_2015_a_2023']
X=df_MinMax_Sem_Tempo.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
#reaplicar o modelo com os melhore hiperparâmetros
modelo_Sem_Tempo = Ridge(alpha=11.12).fit(X,Y)
y_pred = modelo_Sem_Tempo.predict(X)
#cria coluna com as predições
df_MinMax_Sem_Tempo['predicao_melhor_modelo'] = y_pred
mae_total_Sem_Tempo = mean_absolute_error(Y,y_pred)
print("MAE-total", mae_total_Sem_Tempo)
# Calcular os resíduos
df_MinMax_Sem_Tempo['residuos'] =abs(df_MinMax_Sem_Tempo['Recebidos_2015_a_2023'] - df_MinMax_Sem_Tempo['predicao_melhor_modelo'])
df_MinMax_Sem_Tempo['residuos_sinal'] =df_MinMax_Sem_Tempo['Recebidos_2015_a_2023'] - df_MinMax_Sem_Tempo['predicao_melhor_modelo']
print(df_MinMax_Sem_Tempo[['municipio_002','Recebidos_2015_a_2023','predicao_melhor_modelo','residuos_sinal']].sort_values('residuos_sinal').to_string())
print(df_MinMax_Sem_Tempo.columns)
"""

#Cria a coluna 'indice_ordenado'
df_MinMax.sort_values(by='Recebidos_2015_a_2023', ascending=True, inplace = True)
df_MinMax['Recebidos_2015_a_2023']=np.arange(0,df_MinMax.shape[0])
df_MinMax.sort_index(ascending=True, inplace = True)
#colunas = ['Recebidos_2015_a_2023','residuos','predicao_melhor_modelo']
#converte_para_ln(df_MinMax,colunas)
df_MinMax[['Recebidos_2015_a_2023','residuos','predicao_melhor_modelo']]

#plot_residuo(df_MinMax,'indice_ordenado', 'residuos','residuos','predicao_melhor_modelo','residuos')
#esse ultimo 'residuo' na tupla acima diz respeito a algo que nem é usado na funcion. Por isso escrevi qualquer

plotstats(df_MinMax, 'residuos_sinal')


#Tunando Hiperparâmetro
df_MinMax = df_integral.copy()
colunas_a_converter = df_MinMax.dtypes[df_MinMax.dtypes != "object"].index.tolist()
#... exceto a última
colunas_a_converter.pop()
converte_para_MinMax(df_MinMax, colunas_a_converter)
df_MinMax = df_MinMax.drop(colunas_a_converter,axis=1)
#print (df_log.shape)
#df_log.describe()

# Busca pelos melhores hiperparâmetros usando RandomizedSearchCV
# define o modelo
modelo = Ridge()
# Número mínimo de exemplos para fazer o split do nó
alpha_dist = np.linspace(0.01, 100.0, num=10)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
#alphas = [0.1, 0.2, 0.5]
#modelo_randomizado = RandomizedSearchCV(modelo, error_score=mean_absolute_error, param_distributors = random_grid, random_state=33)
modelo_randomizado = RandomizedSearchCV(modelo, {'alpha': alpha_dist}, cv=cv, scoring='neg_mean_absolute_error')
# Roda o modelo de busca randômica.
Y=df_MinMax['Recebidos_2015_a_2023']
X=df_MinMax.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
modelo_randomizado.fit(X, Y)
#x_train = x_log_train
#y_train = y_log_train
#modelo_randomizado.fit(x_train, y_train)
print('MAE: %.3f' % modelo_randomizado.best_score_)
print('Config: %s' % modelo_randomizado.best_params_)

#repetindo o melhor modelo com o melhor hiperparametro
modelo_hyper = Ridge(alpha=11.12, random_state=33)
#Aplicação Ridge() no df_original
resultado = aplicar_modelo_ML('df_MinMax', modelo_hyper, x_minmax_train, y_minmax_train, x_minmax_test, y_minmax_test)
print(resultado)

"""
df_MinMax_Outliers = df_integral_outliers.copy()
#Seleciona as colunas numéricas...
colunas_a_converter = df_MinMax_Outliers.dtypes[df_MinMax_Outliers.dtypes != "object"].index.tolist()
#... exceto a última
colunas_a_converter.pop()
# Chama a função que fará a conversão
converte_para_MinMax(df_MinMax_Outliers, colunas_a_converter)
# Apaga as colunas originais
df_MinMax_Outliers = df_MinMax_Outliers.drop(colunas_a_converter,axis=1)
Y=df_MinMax_Outliers['Recebidos_2015_a_2023']
X=df_MinMax_Outliers.drop(['municipio_002','Recebidos_2015_a_2023'],axis=1)
#reaplicar o modelo com os melhore hiperparâmetros
modelo_outliers = Ridge(alpha=11.12, random_state=33).fit(X,Y)
y_pred = modelo_outliers.predict(X)
#cria coluna com as predições
df_MinMax_Outliers['predicao_melhor_modelo'] = y_pred
mae_total = mean_absolute_error(Y,y_pred)
print("MAE-total", mae_total)
# Calcular os resíduos
df_MinMax_Outliers['residuos'] =abs(df_MinMax_Outliers['Recebidos_2015_a_2023'] - df_MinMax_Outliers['predicao_melhor_modelo'])
df_MinMax_Outliers['residuos_sinal'] =df_MinMax_Outliers['Recebidos_2015_a_2023'] - df_MinMax_Outliers['predicao_melhor_modelo']
print(df_MinMax_Outliers[['municipio_002','Recebidos_2015_a_2023','predicao_melhor_modelo','residuos_sinal']].sort_values('residuos_sinal').to_string())
"""

#LACUNAS
# - Experimentos:
# -   Trazer os outliers de volta
# -   Remover tempo (e outras?)
# -   Em vez de a métrica ser Recebido, ser Recebido/Habitantes 
#       -TRABALHO FUTURO: e esconder a coluna de Habitantes
#       -TRABALHO FUTURO: estimativa fazendo a distancia ser igual para todos
```
